```python
def extract_text_content(html_file):
    file = open(html_file, 'r', encoding = 'utf-8')
    soup = BeautifulSoup(file)        
    for script in soup(["script", "style"]):
        script.decompose()
    strips = list(soup.stripped_strings)
    return strips 
```

```python
def prediction_emails():
    pipeline_validated = load("model/raw_dt.jbl")
    html_files_path = ["emails"+ "/" + x for x in os.listdir(os.getcwd()+"\\emails") if 'html' in x]
    res=[]
    for html_path in html_files_path:
        
        email_texts = extract_text_content(html_path)
        pred = pipeline_validated.predict(email_texts)
        length_total = len(' '.join(email_texts))
        informal_score = 0
        formal_score = 0
        for txt, p in zip(email_texts, pred):
            if p ==0:
                formal_score += len(txt)
            else:
                informal_score+=len(txt)
        informal_score = informal_score/length_total
        formal_score = formal_score/length_total
        
        if informal_score > formal_score:
            print(email_texts[0] + " : Informal")
        else:
            print(email_texts[0] + " : Formal")
        print("Informal score: " + str(informal_score))
        print("Formal score: " + str(formal_score))
        
```

```python
prediction_emails()
```
